```python
import numpy as np
```

```python
scores = dict()

with open('/content/chr21.phyloP46way.placental.wigFix', 'r') as wig_file:
    current_chromosome = ''
    for line in wig_file:
        # start_position = 0
        if line.startswith('fixedStep'):
            current_chromosome = line.split('chrom=')[1].split(' ')[0]  # Extract chromosome name from header
            print(current_chromosome)
            start_position  = int(line.split('start=')[1].split(' ')[0])
            print(start_position)
        else:
            score = line.split('\n')
            score = [float(score[0])]
            if start_position not in scores:
              scores[start_position] = score
            else:
              scores[start_position].append(score)
            # print(score)
            

```

```python
for key, value in scores.items():
    print(f"Length of items for key '{key}': {[len(value)]}")
```
